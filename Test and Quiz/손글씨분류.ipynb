```python
train=pd.read_csv("/kaggle/input/2021-ml-midterm-p3/train.csv")
test=pd.read_csv("/kaggle/input/2021-ml-midterm-p3/test.csv")
submit=pd.read_csv("/kaggle/input/2021-ml-midterm-p3/sample_submit.csv")
```

```python
train
```

```python
x_train=train.drop(["index","label"],axis=1)
y_train=train["label"]
test=test.drop(["index"],axis=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=7).fit(x_train,y_train)
y_pred=clf.predict(test)
submit["label"]=y_pred
```

```python
submit.to_csv("submit.csv",index=False)
```
